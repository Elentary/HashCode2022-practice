```python
def read_file(name):
    with open(name) as f:
        a = [l.strip() for l in f]
        n = int(a[0])
        
        clients = []
        for i in range(n):
            likes = set()
            dislikes = set()
            for j in a[1 + 2 * i].split(' ')[1:]:
                likes.add(j)
            for j in a[1 + 2 * i + 1].split(' ')[1:]:
                dislikes.add(j)
            clients.append
```
